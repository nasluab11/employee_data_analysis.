## 2.Segregate the employees w.r.t different positions

```python
#number of employees in each position
df.groupby('Position')['Name'].count()
```

```python
# Group employees by position
employee_by_position = df.groupby("Position")

for position,employees in employee_by_position:
    print(position)
    print(employees["Name"].to_string(index = 'false'))
```

## 3.Find from which age group most of the employees belong to

```python
# Count employees in each age group
age_count = df['Age'].value_counts(bins = 6)

# Display age group with most employees
max_count_age_group = age_count.idxmax()

print(f"the age group with highest count of employees is {max_count_age_group}")
```

## 4.Find out under which team and position, spending in terms of salary is high

```python
# group by team,poosition and calculate the sum of salary
spending_by_team_position = df.groupby(["Team", "Position"])["Salary"].sum()

# diplaye team-position combination with highest spending
max_spending = spending_by_team_position.idxmax()

print("Team :", max_spending[0])
print("Position :",max_spending[1])
print("Spending :",spending_by_team_position[max_spending])

```

```python
# details of highest paid employee
employee_highest_salary = df["Salary"].idxmax()
df.loc[employee_highest_salary]
```

## 5.Find if there is any correlation between age and salary , represent it visually

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation coefficient between Age and Salary columns
coeff_corr = df["Age"].corr(df["Salary"])

# Plot Age vs. Salary as a scatter plot
plt.scatter(df["Age"],df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title(f"correlation coeffient: {coeff_corr:.2f}")
plt.show()
```
